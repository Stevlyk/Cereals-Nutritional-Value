#### Content
Fields in the dataset:
  ## Name: Name of cereal
  ## mfr: Manufacturer of cereal
* A = American Home Food Products;
* G = General Mills
* K = Kelloggs
* N = Nabisco
* P = Post
* Q = Quaker Oats
* R = Ralston Purina
  ## type:
* C = cold
* H = hot
##### calories: calories per serving
##### protein: grams of protein
##### fat: grams of fat
##### sodium: milligrams of sodium
##### fiber: grams of dietary fiber
##### carbo: grams of complex carbohydrates
##### sugars: grams of sugars
##### potass: milligrams of potassium
##### vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended
##### shelf: display shelf (1, 2, or 3, counting from the floor)
##### weight: weight in ounces of one serving
##### cups: number of cups in one serving
##### rating: a rating of the cereals (Possibly from Consumer Reports?)

```python
# data wrangling library
import pandas as pd

# data visualization library
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
cereals = pd.read_csv('cereal.csv')
```

```python
# ploting a heatmap to show the correlation betweenthe nutritional values and features.

plt.figure(figsize=(12,10))
sns.heatmap(cereals[['calories', 'protein', 'fat', 'sodium', 'fiber',
                    'carbo', 'sugars', 'potass']].corr(), annot=True, cmap='coolwarm')
plt.title('Nutritional Features Correlation')
plt.show
```

# Conclusion:
The correlation analysis reveals that cereals high in calories tends to be high in carbonhydrates and sugar, and low in fiber.
This surggests that many cereals are highly processed and may not provides a balanced nutritional profile. Additionally, cereals high in fat and sodium tends to be high in calories, indicating a potential concern for cardiovascular health.

## Recomendation
Based on this analysis, I recomend to:
* Choose cereals with higher fiber content: Fiber-richcereals tends to be lower in sugar and sodium, this makes it a better choice for balanced diet.
* Be mindful of portion sizes i.e the quantity of intake per time: cereals high in calories, carbonhydrates and sugar can lead to overcunsumption, so it is better to control our intake portions.
* Opt for cereals that has balanced nutritional profiles: look for cereals with a balance of protein, healthy fats and complex carbonhydrates to ensure a nutritious breakfast.
* Check nutrition labels: be aware of the nutrional content of the cereals choices and make informed decisions to support a healthy diet.

```python
# Ploting a scatter plot to show a relationship between calories and sugar content

plt.figure(figsize=(8, 6))
sns.scatterplot(x='calories', y='sugars', data=cereals)
plt.title('Calories vs Sugar Content')
plt.xlabel('Calories')
plt.ylabel('Sugar Content')
plt.show()
```

## Conclusion:
This plot shows that many cereals have a high level of sugar content, which is a concern given the negative health effects of excessive sugar consumption. The correlation between calories and sugar content suggests that cereals with high calories counts are often highly sweetened.

## Recomendation
Based on the analysis, I recommend to:
* Choose cereals with lower sugar content: Opt for cereals with less than 8 grams of sugar per serving to reduce daily sugar intake.
* Be mindful of portion sizes: Even cereals with lower sugar content can lead to excessive sugar consumption if portions are too large.
* Check nutrition labels: Pay attention to both calorie and sugar content when selecting cereals to ensure a balanced choice.
* Consider alternative breakfast options: If finding it challenging to get cereals that meet nutritional standards, consider alternative breakfast options like oatmeal, fruits, or nuts.

```python
# ploting a bar chart to compare cereal ratings by manufacturer

plt.figure(figsize=(12, 6))
sns.barplot(x='mfr', y= 'rating', data=cereals)
plt.title('Cereal Ratings by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Rating')
plt.show()
```

# Conclusion:
This plot shows that consumers have varying preferences when it comes to cereal manufacturers. Nabisco and American Home Food Products followed by Kellogg's are the clear leaders in terms of consumer satisfaction, while other manufacturers lag behind.

## Recomendation
Based on this analysis I recomened to:
* Choose cereals from top-rated manufacturers: Consider cereals from Nabisco and American Home Food Products followed by Kellogg's, as they have consistently high ratings from consumers.
* Explore options from other manufacturers: While Post and Quaker Oats have lower average ratings, they may still offer some popular cereals. General Mills and Ralston Purina may need to improve their product offerings to compete.
* Read reviews and check ingredients: Don't solely rely on manufacturer ratings. Read reviews and check ingredients to ensure the cereal meets your nutritional standards and preferences.

```python
# Ploting a box plot to compare cereal types (hot vs cold) by calories

plt.figure(figsize=(8, 6))
sns.boxplot(x='type', y='calories', data=cereals)
plt.title('Calories by Cereal Type')
plt.xlabel('Type')
plt.ylabel('Calories')
plt.show()
```

# Conclusion
This plot suggests that hot cereals generally have higher calorie counts than cold cereals, indicating a potential concern for consumers watching their calorie intake. However, there is also more variability in the nutritional content of hot cereals, which may offer some healthier options.

## Recomendation
Based on the analysis, I recommend to:
* Choose cold cereals for lower calorie options: If looking to reduce calorie intake, consider choosing cold cereals, which tend to have lower calorie counts.
* Be mindful of portion sizes for hot cereals: If you prefer hot cereals, be aware of the higher calorie content and control your portion sizes to maintain a balanced diet.
* Explore healthier hot cereal options: While hot cereals tend to be higher in calories, some options may be healthier than others. Look for hot cereals with more fiber, protein, and fewer added sugars.
